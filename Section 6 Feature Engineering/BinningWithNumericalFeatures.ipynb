```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('employee.csv')
data.head()
```

```python
data.Age.value_counts().sort_values()
```

```python
def age_cat(x):

  if x>20 and x<30:
    return '20-30'
  elif x>30 and x < 40:
    return '30-40'
  elif x>40 and x<50:
    return '40-50'
  else:
    return '50-60'

data['Age'] = data['Age'].apply(age_cat)
data['Age']
```

```python
data['DistanceFromHome'].value_counts()
```

```python
def distance_cat(x):
  if x<=2:
    return 'Near'
  elif x>2 and x<=5:
    return 'Far'
  else:
    return 'Extremely Far'

data['DistanceFromHome'] = data['DistanceFromHome'].apply(distance_cat)

data['DistanceFromHome'].value_counts()
```

```python
def experience_cat(x):

  if x <= 2:
    return 'Novice'
  elif x > 2 and x <= 5:
    return 'Intermediate'
  return 'experienced'


data['TotalWorkingYears'] = data['TotalWorkingYears'].apply(experience_cat)

data['TotalWorkingYears'].value_counts()
```
