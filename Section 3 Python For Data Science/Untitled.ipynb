# Merging DataFrames

```python
data1=pd.read_csv('market_fact.csv')
data2=pd.read_csv('orders_dimen.csv')
```

```python
data1.head()
```

```python
data2.head()
```

```python
cust_order = pd.merge(data1,data2,on='Ord_id')
```

```python
cust_order.head()
```

<strong>NB:</strong> We can specify the way we want to join the two dataframes with the <strong>how</strong> parameter in the <strong>pd.merge()</strong> function.

We have 4 ways to join two tables:
<ul>
    <li><strong>Inner join: </strong>it combines two dataframes based on common key and return a new dataframe that contains only rows that have a matching values in both original dataframes.</li>
    <li><strong>outer join: </strong>it returns all those records which either have a match in the left or in the right dataframe.</li>
    <li><strong>Left join: </strong>it returns a dataframe containing all rows of the left dataframe.All the non-matching rows of the left dataframe contain <strong>NaN</strong> for the columns in the right dataframe.</li>
    <li><strong>Right join: </strong> it returns a dataframe containing all rows of the right dataframe.All the non-matching rows of the right dataframe contain <strong>NaN</strong> for the columns in the left dataframe</li>
</ul>


```python
cust_order = pd.merge(data1,data2,on='Ord_id',how='inner')
cust_order
```

# Concatenation

```python
# dataframes having the same columns
df1 = pd.DataFrame({'Name': ['Arijit', 'Neeraj', 'Sakshi', 'Muskan'],
                    'Age': [34, 30, 22, 33],
                    'Gender': ['M', 'M', 'F', 'M']}
                  )

df2 = pd.DataFrame({'Name': ['Kartik', 'Veer', 'Preeti'],
                    'Age': [31, 22, 19],
                    'Gender': ['M', 'M', 'F']}
                  )
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1, df2], axis = 0)
df= df.reset_index()
df = df.drop(['index'], axis = 1)
df
```

```python
df = df1.append(df2)
df = df.reset_index()
df = df.drop(['index'], axis = 1)
df
```

#  Lambda, Map, and Apply Functions

```python
data=pd.read_csv('employees.csv')
```

```python
data.head()
```

```python
# lets define a function to divide the Daily Rate by 100
def function(x):
    return x/10

data['SALARY'] = data['SALARY'].map(function)
data['SALARY']
```

<strong>NB: map()</strong> function can be used only with series not dataframe.

```python
data[['SALARY', 'DEPARTMENT_ID']].map(function)
```

## Apply function

<strong>NB: apply()</strong> function is can alternative for <strong>map()</strong> function. It can be used with dataframe.

```python
data[['SALARY', 'DEPARTMENT_ID']].apply(function)
```

## Lambda function

```python
data['SALARY'] = data['SALARY'].apply(lambda x: x/10)
print(data['SALARY'])
```
